package main

import (
	"fmt"
)

func calcularRendaTotal(ingressos int, popular, geral, arquibancada, cadeiras float64) float64 {
	// Calcula o número de ingressos vendidos para cada categoria
	ingressosPopular := float64(ingressos) * (popular / 100.0)
	ingressosGeral := float64(ingressos) * (geral / 100.0)
	ingressosArquibancada := float64(ingressos) * (arquibancada / 100.0)
	ingressosCadeiras := float64(ingressos) * (cadeiras / 100.0)

	// Calcula a renda total do jogo
	rendaTotal := (ingressosPopular * 1.0) + (ingressosGeral * 5.0) + (ingressosArquibancada * 10.0) + (ingressosCadeiras * 20.0)
	return rendaTotal
}

func main() {
	var numCasos int
	fmt.Scanln(&numCasos)

	for i := 1; i <= numCasos; i++ {
		var ingressos int
		var popular, geral, arquibancada, cadeiras float64
		fmt.Scanln(&ingressos, &popular, &geral, &arquibancada, &cadeiras)

		renda := calcularRendaTotal(ingressos, popular, geral, arquibancada, cadeiras)
		fmt.Printf("A RENDA DO JOGO N. %d E = %.2f\n", i, renda)
	}
}

package main

import (
	"fmt"
	"strconv"
	"strings"
)

func main() {
	var n1, n2, n3 int
	fmt.Println("Digite três números inteiros separados por espaço:")
	_, err := fmt.Scanln(&n1, &n2, &n3)
	if err != nil {
		fmt.Println("Erro ao ler os números:", err)
		return
	}

	// Verifica se algum dos números tem mais de um dígito
	if n1 < 0 || n1 > 9 || n2 < 0 || n2 > 9 || n3 < 0 || n3 > 9 {
		fmt.Println("DIGITO INVALIDO")
		return
	}

	// Concatena os três números para formar o número inteiro
	numero := strconv.Itoa(n1) + strconv.Itoa(n2) + strconv.Itoa(n3)

	// Calcula o quadrado do número
	quadrado := (n1*100 + n2*10 + n3) * (n1*100 + n2*10 + n3)

	// Remove zeros à esquerda do número, se houver
	numero = strings.TrimLeft(numero, "0")

	// Imprime o número resultante e seu quadrado
	fmt.Printf("%s, %d\n", numero, quadrado)
}

package main

import "fmt"

func main() {
    var conta int
    var consumo float64
    var tipoConsumidor string

    // Solicita ao usuário que insira a conta do cliente, o consumo de água e o tipo de consumidor
    fmt.Println("Digite a conta do cliente, o consumo de água (metros cúbicos) e o tipo de consumidor (R - Residencial, C - Comercial, I - Industrial), separados por espaço:")
    fmt.Scan(&conta, &consumo, &tipoConsumidor)

    // Calcula o valor da conta de acordo com o tipo de consumidor
    var valorConta float64

    switch tipoConsumidor {
    case "R":
        valorConta = 5.00 + (0.05 * consumo)
    case "C":
        if consumo <= 80 {
            valorConta = 500.00
        } else {
            valorConta = 500.00 + (0.25 * (consumo - 80))
        }
    case "I":
        if consumo <= 100 {
            valorConta = 800.00
        } else {
            valorConta = 800.00 + (0.04 * (consumo - 100))
        }
    default:
        fmt.Println("Tipo de consumidor inválido.")
        return
    }

    // Imprime a conta do cliente e o valor da conta
    fmt.Printf("CONTA = %d\n", conta)
    fmt.Printf("VALOR DA CONTA = %.2f\n", valorConta)
}

func calcularValorAluguel(horas float64) float64 {
	var valorTotal float64

	// Verificar quantas vezes o pacote de 3 horas foi completado
	pacotesDeTresHoras := int(horas / 3)
	valorTotal += float64(pacotesDeTresHoras) * 10.0

	// Calcular o valor para as horas restantes, se houver
	horasRestantes := horas - float64(pacotesDeTresHoras*3)
	if horasRestantes > 0 {
		valorTotal += horasRestantes * 5.0
	}

	return valorTotal

package main

import "fmt"

func main() {
    var a1, r, n int

    // Solicita os valores do usuário
    fmt.Println("Digite o valor inicial da progressão, a razão e o número de termos separados por espaço:")
    fmt.Scan(&a1, &r, &n)

    // Calcula a soma dos n primeiros termos da progressão
    soma := calcularSomaProgressaoAritmetica(a1, r, n)

    // Imprime o resultado
    fmt.Println("A soma dos", n, "primeiros termos da progressão aritmética é:", soma)
}

// Função para calcular a soma dos n primeiros termos de uma progressão aritmética
func calcularSomaProgressaoAritmetica(a1, r, n int) int {
    soma := 0
    for i := 0; i < n; i++ {
        termo := a1 + i*r
        soma += termo
    }
    return soma
}

package main

import (
	"fmt"
)

func main() {
	var n int

	// Solicita o número inteiro positivo ao usuário
	fmt.Println("Digite um número inteiro positivo maior que 1:")
	_, err := fmt.Scan(&n)
	if err != nil || n <= 1 {
		fmt.Println("Número inválido!")
		return
	}

	// Calcula a soma da série
	soma := calcularSerie(n)

	// Imprime o resultado com 6 casas decimais
	fmt.Printf("%.6f\n", soma)
}

// Função para calcular a soma da série
func calcularSerie(n int) float64 {
	soma := 0.0
	for k := 1; k <= n; k++ {
		soma += 1.0 / float64(k)
	}
	return soma
}
